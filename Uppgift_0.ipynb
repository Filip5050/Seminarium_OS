### A. Antal länder som har varit med i OS genom åren


```python
import pandas as pd

# läs CSV filen
os = pd.read_csv('athlete_events.csv')

# Räkna antal unika länder 
antal_länder = os['NOC'].nunique()

# Visa antal länder 
print(f'Det är {antal_länder} länder som har varit med i OS genom åren.')
```

---
### B. Länder som har spelat i OS genom åren



```python
# Tabulate delen av koden nedan är hämtad från https://www.packetswitch.co.uk, skriven av Suresh Vina
# Källa: https://www.packetswitch.co.uk/how-to-create-tables-python-tabulate-module/

from tabulate import tabulate

# Hämta alla unika länder
länder = os['NOC'].dropna().unique()

# Skapa en DataFrame för att visa länderna i tabellformat
unika_länder = pd.DataFrame(länder, columns=['Land'])

# Lägg till en nummerkolumn
unika_länder.insert(0, 'Nummer', range(1, len(unika_länder) + 1))

# Använd tabulate för att skriva ut tabellen  
print(tabulate(unika_länder, headers='keys', tablefmt='pretty', showindex=False))
```

---
### C. Sporter som har spelats i OS genom åren

```python
# Hämta unika värden från kolumnen "Sport"
sporter = os["Sport"].dropna().unique()

# Skapa en DataFrame från de unika värdena
unik_sport = pd.DataFrame(sporter, columns=["Sporter"])

# Lägg till en kolumn med nummer
unik_sport["Nummer"] = range(1, len(unik_sport) + 1)

# Omvandla kolumner så att "Nummer" kommer först
unik_sport = unik_sport[["Nummer", "Sporter"]]

# Använd tabulate för att skriva ut tabellet
print(tabulate(unik_sport, headers="keys", tablefmt="grid", showindex=False))
```

### F. Antal medaljer som Frankrike har vunnit genom åren

```python
import matplotlib.pyplot as plt

# Filtrera data och räkna medaljer per år
frankrike_medals = os[os['Team'] == 'France'].groupby('Year')['Medal'].value_counts().unstack().fillna(0)

# Välj dessa kolumner
frankrike_medals = frankrike_medals[['Gold', 'Silver', 'Bronze']]

print(frankrike_medals)

# Skapa en graf
frankrike_medals.plot(kind='bar', stacked=True, figsize=(8, 6))

# Lägg till titlar och etiketter för grafen
plt.title('Antal medaljer per år för Frankrike i OS', fontsize=16, fontweight= 'bold')
plt.xlabel('År', fontsize=12, fontweight = 'bold')
plt.ylabel('Antal medaljer', fontsize=12, fontweight='bold')
plt.xticks(rotation=45)  # Rotera årtalen för bättre läsbarhet
plt.legend(title='Medalj', fontsize=10)
plt.tight_layout()  # För att undvika att text klipper av vid kanterna

plt.show()
```

### G. Diagram över könsfördelningen i OS genom åren

```python

# Räkna antalet manliga och kvinnliga deltagare
kön = os['Sex'].value_counts()

# Skapa ett stapeldiagram
kön.plot(kind='bar', color=['blue', 'green'], figsize=(6, 4))

# Lägg till titlar och etiketter
plt.title('Könsfördelning bland deltagare i OS', fontsize=14, fontweight='bold')
plt.xlabel('Kön', fontsize=12, fontweight = "bold")
plt.ylabel('Antal Deltagare', fontsize=12, fontweight = 'bold')
plt.xticks(rotation=0)  # För att hålla etiketterna horisontella

plt.tight_layout()
plt.show()
```

### H. Diagram över topp 10 länder som tagit flest medaljer genom åren

```python
# Räkna antalet medaljer per land
medalj_per_land = os.groupby('NOC')['Medal'].count()

# Hämta de top 10 länderna med flest medaljer
top_10_länder = medalj_per_land.sort_values(ascending=False).head(10)

# Skapa ett diagram för de top 10 länderna
plt.figure(figsize=(6, 4))
top_10_länder.plot(kind='bar', color='orange')

# Sätt etiketter och titel för diagrammet
plt.ylabel('Antal Medaljer', fontsize = 10, fontweight = 'bold')
plt.xlabel('Land', fontsize = 10, fontweight = 'bold')
plt.title('Top 10 Länder med Flest Medaljer i OS', fontsize = 15, fontweight = 'bold')
plt.xticks(rotation=0)
# Justera layout så att etiketter och diagram inte överlappar
plt.tight_layout()

plt.show()
```

## Några intresanta data från OS genom åren

### 1. Medaljfördelning per kön

```python
# Medaljfördelning per kön
medaljer_per_kon = os.groupby(['Sex', 'Medal']).size().unstack(fill_value=0)

# Plotta ett stackat stapeldiagram
medaljer_per_kon.plot(kind='bar', stacked=True, figsize=(10, 6), color=[ 'brown', 'gold', 'silver'])
plt.title('Medaljfördelning per kön', fontsize=16, fontweight='bold')
plt.xlabel('Kön', fontsize=12)
plt.ylabel('Antal medaljer', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

### 2. Antal medaljer i Sommar-OS och Vinter-OS

```python
# Antal medaljer i Sommar-OS och Vinter-OS
medaljer_per_season = os.groupby(['Season', 'Medal']).size().unstack(fill_value=0)

# Plotta ett stapeldiagram
medaljer_per_season.plot(kind='bar', stacked=True, figsize=(8, 6), color=['red', 'gold' , 'silver'])
plt.title('Medaljfördelning per säsong (Sommar vs Vinter)', fontsize=16, fontweight='bold')
plt.xlabel('Säsong', fontsize=12)
plt.ylabel('Antal medaljer', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

### 3. Antal medaljer per år i OS genom åren

```python
# Räkna medaljer per år
medaljer_per_år = os.groupby('Year')['Medal'].count()

# Plotta grafen
plt.figure(figsize=(10,6))
medaljer_per_år.plot(kind='bar', color='purple')
plt.title('Antal medaljer per år i OS', fontsize=16, fontweight='bold')
plt.xlabel('År', fontsize=12)
plt.ylabel('Antal medaljer', fontsize=12)
plt.show()
```

### 4. Antal deltagare per år i OS

```python
# Antal deltagare per år
deltagare_per_år = os.groupby('Year')['ID'].nunique()

# Plotta linjediagram
plt.figure(figsize=(10,6))
plt.plot(deltagare_per_år.index, deltagare_per_år, marker='o', color='green')
plt.title('Antal deltagare per år i OS', fontsize=16, fontweight='bold')
plt.xlabel('År', fontsize=12)
plt.ylabel('Antal deltagare', fontsize=12)
plt.grid(True)
plt.show()
```
